```python
# rename some of the header for better spacing
data = data.rename(columns={'Height (Inches)':'Height/inches', 
                        'Weight (Pounds)':'Weight/lbs',
                        'Grip strength':'Grip/strength'})
data
```

```python
import matplotlib.pyplot as plt

# show the average grip strength by frailty of each person
rate = data.groupby(['Frailty'])['Grip/strength'].mean()
print(rate)

# figure size
fig = plt.figure(figsize =(10, 7))

# display 
plt.bar(['No','Yes'], rate, color=['orange', 'pink'], width = 0.4)
plt.xlabel("Pass or Gone")
plt.ylabel("Strength Number")
plt.title("Grip strength by Frailty")
plt.show()
```

```python
import matplotlib.pyplot as plt

myexplode = [0.1, 0]

# Yes:No Ratio in Frailty
rate = data.groupby(['Frailty'])['Frailty'].count()
print(rate)

# figure size
fig = plt.figure(figsize =(10, 7))

# display
plt.pie(rate, autopct='%1.0f%%', explode=myexplode, shadow=True)
plt.legend(labels=['No','Yes'])
plt.title("Yes:No Ratio in Frailty")
plt.show()
```
